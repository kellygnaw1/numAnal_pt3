To logically solve this problem, let us first write edown the explicit finite difference for $(6)$ in the question, find the order of the method, and give stability conditions using Von Neumann analysis.<br/><br/>

We can use the central differencing for the second derivative and let $a = 1$ and $\lambda = \frac{k}{h}$. 

$$\frac{p_j^{n+1} - 2p_j^n + p_j^{n-1}}{k^2} = \frac{p_{j+1}^n - 2p_j^n + p_{j-1}^n}{h^2}\quad \quad \text{explicit finite difference}\\
p_j^{n+1} = 2p_j^n - p_j^{n-1} + \lambda^2\left(p_{j+1}^n - 2p_j^n + p_{j-1}^n\right)\\
= \lambda^2 \left(p_{j+1}^n + p_{j-1}^n\right) + 2(1-\lambda^2)p_j^n - p_j^{n-1}
$$

Now apply taylor approximation at  $(t_n, x_j)$ 
$$
\begin{align}
p(t_n+k, x_j) &= p(t_n, x_j) + k\frac{\partial p}{\partial t}(t_n, x_j) + \frac{k^2}{2}\frac{\partial^2 p}{\partial^2 t} + \dots\\
p(t_n - k, x_j) &= p(t_n, x_j) - k\frac{\partial p}{\partial t}(t_n, x_j) + \frac{k^2}{2}\frac{\partial^2 p}{\partial^2 t} - \dots\\
p(t_n, x_j + h) & = p(x_n, x_j) + h\frac{\partial p}{\partial x}(t_n, x_j) + \frac{h^2}{2}\frac{\partial^2 p}{\partial^2 x}(t_n, x_j) + \dots\\
p(t_n, x_j - h) & = p(t_n, x_j) - h\frac{\partial p}{\partial x}(t_n, x_j) + \frac{h^2}{2}\frac{\partial^2 p}{\partial^2 x}(t_n, x_j) - \dots\\
\end{align}
$$
Let $p_j^{n+1} = p(t_{n+1}, x_j), p_j^n = p(t_n, x_j), p_j^{n-1} = p(t_{n-1}, x_j), p_{j+1}^n = p(t_n, x_{j+1}), p_{j-1}^n = p(t_n, x_{j-1})$, we can substitute this into the explicit finite difference using the taylor expansion from above. 
$$
\begin{align}
\frac{p_j^{n+1} - 2p_j^n + p_j^{n-1}}{k^2} &= \frac{1}{k^2}\left(p + k\frac{\partial p}{\partial t} + \frac{k^2}{2}\frac{\partial^2 p}{\partial^2 t} + \frac{k^3}{3!}\frac{\partial^3 p}{\partial^3 t}+\dots -2p + p - k\frac{\partial p}{\partial t} + \frac{k^2}{2}\frac{\partial^2 p}{\partial^2 t} - \frac{k^3}{3!}\frac{\partial^3 p}{\partial^3 t}+... \right)\\
&= \frac{1}{k^2}\left(k^2\frac{\partial^2 p}{\partial^2 t} + \frac{k^4}{\frac{4!}{2}}\frac{\partial^4 p}{\partial^4 t} + \dots\right)\\
&= \frac{\partial^2 p}{\partial^2 t} + \frac{k^2}{\frac{4!}{2}}\frac{\partial^4 p}{\partial^4 t}+\dots
\end{align}
$$
<br/><br/>
$$\begin{align}
\frac{p_{j+1}^n - 2p_j^n + p_{j-1}^n}{h^2} &= \frac{1}{h^2}\left(p + h\frac{\partial p}{\partial x} + \frac{h^2}{2}\frac{\partial^2 p}{\partial^2 x} + \frac{h^3}{3!}\frac{\partial^3 p}{\partial^3 x}+\dots -2p + p - h\frac{\partial p}{\partial x} + \frac{h^2}{2}\frac{\partial^2 p}{\partial^2 x} - \frac{h^3}{3!}\frac{\partial^3 p}{\partial^3 x}+... \right)\\
&= \frac{1}{h^2}\left(h^2\frac{\partial^2 p}{\partial^2x} + \frac{h^4}{\frac{4!}{2}}\frac{\partial^4 p}{\partial^4 x} + \dots\right)\\
&= \frac{\partial^2 p}{\partial^2 x} + \frac{h^2}{\frac{4!}{2}}\frac{\partial^4 p}{\partial^4 x}+\dots
\end{align}$$
<br/><br/>
Since $a = 1, \frac{\partial^2p}{\partial^2x} = \frac{\partial^2p}{\partial^2t}$ so then we can see the following:
$$
\begin{align}
\frac{p_j^{n+1} - 2p_j^n + p_j^{n-1}}{k^2} - \frac{p_{j+1}^n - 2p_j^n + p_{j-1}^n}{h^2} &= \\
&\frac{\partial^2 p}{\partial^2 t} + \frac{k^2}{\frac{4!}{2}}\frac{\partial^4 p}{\partial^4 x}+\dots -\left(\frac{\partial^2 p}{\partial x}  + \frac{h^2}{\frac{4!}{2}} \frac{\partial^4 p}{\partial^4 x} + \dots \right)=\\
&O(k^2) + O(h^2)
\end{align}
$$

<br/><br/>
Now to find the stability condition, let $p_j^n = g^ne^{imjh}:$ (Note this will be similar to homework 4 so I skipped some steps)
$$\begin{align}
\frac{1}{k^2}\left(g^{n+1}e^{imjh} - 2g^ne^{imjh} + g^{n-1}e^{imjh}\right) &= \frac{1}{h^2}\left(g^ne^{im(j-1)h} - 2g^ne^{imjh} + g^ne^{im(j+1)h}\right)\\
\frac{1}{k^2}\left(g-2+\frac{1}{g}\right) &= \frac{1}{h^2}\left(e^{-imh} - 2 + e^{imh}\right)\\
g-2+\frac{1}{g} &= -4\lambda^2(\sin^2(\frac{mh}{2}))\\
(g^{.5} - g^{.5})^2 &= \pm(-4\lambda^2(\sin^2(\frac{mh}{2})))\\
(g^{.5} - g^{.5}) &= \pm(2\lambda(i\sin(\frac{mh}{2}))\\
g-1 &= \pm 2ig^{1/2}\lambda\sin(\frac{mh}{2})\\
0 &= g\pm2ig^{1/2}\lambda\sin(\frac{mh}{2})+1\\
\text{(Apply quadratic formula)} \quad g &= \left(\sqrt{1-\lambda^2\sin^2(\frac{mh}{2})} \pm i\lambda\sin(\frac{mh}{2})\right)
\end{align}
$$

<br/><br/>
From above, we can see that the first term is real if $|\lambda| \leq 1$ so this finite difference scheme is stable for $|\lambda| \leq 1$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#code for 3b
#initial function 
def f(x, p):
    u = np.zeros(len(x))
    for i in range(len(x)):
        if i == 0 or i == len(x)-1:
            u[i] = p
        u[i] = p * np.cos(2*x[i] * np.pi)
    return u

#define scheme 
def open_FDScheme(A, B, N):
    u = np.zeros(N+1)
    for i in range(N):
        if i == 0:
            u[i]  = A[0]
            u[-1] = B[-1]
        else:
            u[i] = A[i+1] + A[i-1] - B[i]
    return u
```

```python
#plot for 3b
"""
L        = pipe, time length 
p0       = amplitutde 
delta_xt = delta x and delta t 
N        = space steps and time steps 
"""
L        = 1
p0       = 0.9
delta_xt = 0.05
N        = int(L/delta_xt)
u        = np.zeros((N+1, N+1))
x        = [i * delta_xt for i in range(N+1)]
u[0]     = f(x, p0)
u[1]     = u[0]
for t in range(2, N+1):
    u[t] = open_FDScheme(u[t-1], u[t-2], N)

print('--------------------------Plots for Open Pipe--------------------------')
plt.plot(x, u[10])
plt.title("t = 0.5")
plt.show()

plt.plot(x, u[20])
plt.title("t = 1.0")
plt.show()
```

```python
#code for 3c
def close_FDScheme(A, B, N):
    u = np.zeros(N+1)
    for i in range(N):
        if i == 0:
            u[i]  = A[0]
            u[-1] = 2 * A[-2] - B[-1]
        else:
            u[i] = A[i+1] + A[i-1] - B[i]
    return u
```

```python
#plot for 3c
for i in range(2, N+1):
    u[i] = close_FDScheme(u[i-1], u[i-2], N)
    
print('--------------------------Plots for Closed Pipe--------------------------')
plt.plot(x, u[10])
plt.title("t = 0.5")
plt.show()

plt.plot(x, u[20])
plt.title("t = 1.0")
plt.show()    
```
